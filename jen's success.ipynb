```python
import numpy as np
import pylab as pl

# Use numpy to load the data contained in the files into a 2-d array called data
x1 = np.loadtxt('LCO_Bavg')


#Plot the data.  the - 2457999.12 is what we define as day 0.  
#   You can subtract any number depending on what your data is in to shift the lightcurve on the x axis.  
#   For the y axis I've subtracted the extinction (1.1) and the distance modulus (31.46) so that it is in absolute magnitude.  
#   You can also change the marker type and the color.  Just google "pyton marker" and it'll give you a list of things!
pl.plot(x1[:,0]-2457999.12, x1[:,1]-1.1-31.46,color='blue',marker='o',linestyle='',markersize=9,fillstyle='none' )


#If you want to plot a line between some points:
pl.plot([-2,200], [-12.06,-12.06],linestyle='--', color='black', linewidth=2)


#SLphot - if all of the data is in one text file and you want to plot them with error bars.  
#Remember, in python the first column is 0 (not 1)
x7 = np.loadtxt('SLPhot')

pl.errorbar(x7[:,0]-2457999.12, x7[:,1]-1.1-31.46+.11, yerr=x7[:,2],color='blue',marker='o',linestyle='',markersize=9,label='B')
pl.errorbar(x7[:,0]-2457999.12, x7[:,3]-0.8-31.46-.13, yerr=x7[:,4],color='green',marker='h',linestyle='',markersize=9, label='V')
pl.errorbar(x7[:,0]-2457999.12, x7[:,5]-0.63-31.46, yerr=x7[:,6],color='#ff5722',marker='v',linestyle='',markersize=9, label='R')
pl.errorbar(x7[:,0]-2457999.12, x7[:,7]-0.44-31.46, yerr=x7[:,8],color='#e51c23',marker='*',linestyle='',markersize=11, label='I')

#labels for the plots
pl.xlabel('Days after Explosion',fontsize='18')
pl.ylabel('Absolute Magnitude',fontsize='18')

#if you want to add any text
pl.text(173,-15.3,'$^{56}$Co',fontsize='12', color='darkgray')

#you can change the x and y axis ranges here:
pl.xlim(-5,190)
pl.ylim(-13.7,-19.3)

#if you want a legend
pl.legend(ncol=2)

#if you want to save the file, delete the # sign
#pl.savefig('/Users/Hermione/Research/SNe/SN2017gmr/FullLC.eps')
pl.show()
```
